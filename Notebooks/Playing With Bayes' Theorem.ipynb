```python
# Generating some simple photon count data
import numpy as np
from scipy import stats
np.random.seed(1)  # for repeatability
%matplotlib inline
import matplotlib.pyplot as plt

F_true = 1000  # true flux, say number of photons measured in 1 second
N = 50 # number of measurements
F = stats.poisson(F_true).rvs(N)  # N measurements of the flux
e = np.sqrt(F)  # errors on Poisson counts estimated via square root

fig, ax = plt.subplots()
ax.errorbar(F, np.arange(N), xerr=e, fmt='ok', ecolor='gray', alpha=0.5)
ax.vlines([F_true], 0, N, linewidth=5, alpha=0.2)
ax.set_xlabel("Flux");ax.set_ylabel("measurement number");
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt

coinbiased=random.randrange(0,100)/100 #probability of coin biased towards heads
bias=random.randrange(50,100)/100
print(coinbiased,'chance of coin being biased with bias:', bias)
```

```python
def randomflip():
    flip=pow(random.randrange(0,100)/100, 0.3)
    if flip>0.5:
        return 'heads'
    if flip<0.5:
        return 'tails'
```

```python
coinbiased=0.75
bias=0.50001
nonbiased=0.5
flipcatalog=[]
posteriorcatalog=[coinbiased]
icatalog=[0]
posterior=coinbiased
for i in range(1000000):
    flip=randomflip()
    flipcatalog.append(flip)
    if flip=='heads':
        posterior=(bias*posterior)/(bias*posterior+nonbiased*(1-posterior))
    if flip=='tails':
        posterior=((1-bias)*posterior)/((1-bias)*posterior+(1-nonbiased)*(1-posterior))
    posteriorcatalog.append(posterior)
    icatalog.append(i)

print(posterior, 'probability of coin being biased towards heads with a coinbiased chance', coinbiased, 'and bias', bias, 'given the following sequence of flips')
fig = plt.figure()

ax = plt.gca()
ax.plot(icatalog, posteriorcatalog,color='green', marker='o', linestyle='dashed', linewidth=2, markersize=5)
fig = plt.gcf()
fig.set_size_inches(20, 12)
```

```python
heads=0
tails=0
for i in range(100000):
    flip2=randomflip()
    if flip2=='heads':
        heads+=1
    if flip2=='tails':
        tails+=1
print(tails,heads)
```
